$Q = C * H_{a}^{n} \tag{1}$

$\frac{H_a}{W} = [0.4,0.5,0.6,0.7,0.8]$

$W:$ancho de graganta (cm) <br>
$Ha:$Profundidad del agua en la posición a (pies, según ecuación 1) <br>
$C:$Coefiente que depende de la dimensión del canal <br>
$n:$Coefiente que depende de la dimensión del canal <br>
$Q:$Caudal (pies cúbicos/seg, según ecuación 1)

$1 cm = 0.032808 ft$

$1 \frac{ft^3}{s}= 0.0283 \frac{m^3}{s}$

```python
HaW = np.array([0.4,0.5,0.6,0.7,0.8])
```

```python
for i in range(len(HaW)):
  Ha = W*0.032808*HaW[i]
  Q = C*Ha * 0.0283
  plt.plot(Q, Ha/3.2808, label = HaW[i])
  plt.legend(title="Ha/W")
  plt.xlabel("Q, $m^3$/s")
  plt.ylabel("Ha, m")
```

```python
for i in range(len(HaW)):
  Ha = W*0.032808*HaW[i]
  Q = C*Ha * 0.0283
  plt.plot(Q, Ha/3.2808, label = HaW[i])
  plt.legend(title="Ha/W")
  plt.xlabel("Q, $m^3$/s")
  plt.ylabel("Ha, m")
  plt.xlim(0, 0.5)
  plt.ylim(0,0.6)
```

```python
for i in range(len(HaW)):
  Ha = W*0.032808*HaW[i]
  Hb = Ha*HbHa
  Q = C*Ha * 0.0283
  plt.plot(Q, Hb/3.2808, label = HaW[i])
  plt.legend(title="Ha/W")
  plt.xlabel("Q, $m^3$/s")
  plt.ylabel("Hb, m")
  plt.xlim(0, 0.5)
  plt.ylim(0,0.6)
```
