# Conjunto de Mandelbrot

<p> Este conjunto se define en el plano complejo fijando un número complejo c cualquiera. A partir de c, se construye una sucesión por recursión:</p><br>

$$
\begin{equation}
f_{c}(z) =
    \begin{cases}
        z_{0} =& 0\in \mathbb{C} \hspace{1mm}(Término \hspace{1mm} inicial)\\
        z_{n+1} =& z_{n}^{2} + c \hspace{1mm}(Sucesión \hspace{1mm} recursiva)
    \end{cases}
\end{equation}
$$ <br>

<p>Si esta sucesión queda acotada, entonces se dice que c pertenece al conjunto de Mandelbrot, y si no, queda excluido del mismo.</p>
